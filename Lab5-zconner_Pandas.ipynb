### Run calculations on pandas dataframes

```python
# convert temperature from fahrenheit to celsius using 
# C = (F - 32)/ 1.8

centralparktemp.iloc[:, 1:14] = ((centralparktemp.iloc[:, 1:14] - 32) / 1.8)
centralparktemp
```

```python
# convert precipitation from in to mm using mm = in * 25.4
centralparkprecip.iloc[:, 1:14] = ((centralparkprecip.iloc[:, 1:14] * 25.4))
centralparkprecip
```

### Calculate summary statistics of pandas dataframes

```python
# use .describe() to get summary statistics
centralparktemp.describe()
```

```python
# plot a box plot of the precipitation
centralparkprecip.iloc[:,1:13].boxplot(whis = 'range')

# range was 1:13 because the annual (14) skewed the data
```

```python
# use .describe() to get summary statistics
centralparktemp.describe()
```

```python
# plot a box plot of the temperature
centralparktemp.iloc[:,1:13].boxplot(whis = 'range')
```

### Plot boxplot of precip and temp

```python
# plot the boxplot of monthly and annual precip and temp as two subplots
# create the plot area
fig, (ax1, ax2) = plt.subplots(2, figsize = (15, 15))

# add the x and y axes for temperature
centralparktemp.iloc[:,1:13].boxplot(ax = ax1, whis = 'range')

# set plot title
ax1.set(title = '1869 - 2017 Monthly Temperature in Central Park, NY')

# add labels to the axes
ax1.set(ylabel = 'Temperaute(C)')

# add the x and y axes for precipitation
centralparkprecip.iloc[:,1:13].boxplot(ax = ax2, whis = 'range')

# set plot title
ax2.set(title = '1869 - 2017 Monthly Precipitation in Central Park, NY')

# add labels to the axes
ax2.set(ylabel = 'Precipitation(mm)')

```

### Numpy and Pandas dataframes comparison

Pandas is similar to numpy in that you are able to create 2D and 3D arrays of data. A difference between the pandas and the numpy arrays, is that numpy arrays must contain the same data type throughout. On the other hand, with pandas dataframes, you can have a different datatype in each column that can be imported from CSV, JSON or from dicitionaries. 

When it came to converting the data from inches to millimeters, I found little difference in the difficulty between the two packages.
